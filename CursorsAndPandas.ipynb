# Using Cursors, Dictionaries and Pandas

```python
import arcpy
```

```python
fc = "C:\EsriTraining\PYTS\Data\CountyData.gdb\ParcelPts"
```

```python
sFields = ['Owner_Name', 'Street_Name', 'Street_Number', 'City', 'State', 'SHAPE@XY']
```

```python
dct = {}
```

```python
with arcpy.da.SearchCursor(fc, sFields) as sCursor:
    for row in sCursor:
        #print(row[0])
        dct[row[0]] = { 
                        'Address' : "{} {}, {}".format(row[1],
                                                       row[2],
                                                       row[3]),
                        'State' : row[4],
                        'Street' : row[1],
                        'StreetNumber' : row[2],
                        'Geometry' : row[-1]
                      }
```
